```python
import matplotlib.pyplot as plt
import calmap
import pandas as pd 
```

```python
# import Yahoo Stock Data
# by using parse_dates, we're telling pandas to parse columns as datetime
df = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/yahoo.csv", parse_dates=['date'])

#set date as index
df.set_index('date', inplace=True)
```

```python
#plot the gride and specify dimensions and pixels
plt.figure(figsize=(16,10), dpi= 80)

#call calmap and tell it to plot 2008 
#syntax: calmap.calendarplot(data, how=u'sum', yearlabels=True, yearascending=True, yearlabel_kws=None, subplot_kws=None, gridspec_kws=None, fig_kws=None, **kwargs)[source]

#here we're telling calendarplot to visualize the VIX 
calmap.calendarplot(df['2008']['VIX.Close'], fig_kws={'figsize': (16,10)}, yearlabel_kws={'color':'black', 'fontsize':14}, subplot_kws={'title':'Yahoo Stock Prices'})
plt.show()


```
